```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import numpy as np
import pandas as pd

iris = pd.read_csv("iris.csv")
iris.head()
```

```python
np.random.seed(0)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris.drop(["Species"], axis=1), iris["Species"], test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf  = RandomForestClassifier(n_estimators=116)

rf.fit(X_train, y_train)

print("Genauigkeit der Testdaten: {:.4f}".format(rf.score(X_test, y_test)))
```
